```python
import pybullet as p
import pybullet_data as pd
import time

# open the GUI
p.connect(p.GUI)

def setupWorld():
	p.resetSimulation()
	# p.setPhysicsEngineParameter(deterministicOverlappingPairs=1)
	p.setAdditionalSearchPath(pd.getDataPath())
	
	# load files and place them at the offsets
	print("turtle")
	turtle = p.loadURDF("urdf/most_simple_turtle.urdf",[0,0,1])
	print("plane")
	plane = p.loadURDF('plane100.urdf')
	# plane = p.loadURDF("urdf/plane_box.urdf")
	print("box1")
	box1 = p.loadURDF("urdf/box.urdf", [1,0,1])
	print("box2")
	box2 = p.loadURDF("urdf/box.urdf", [1,1,1])

	# define gravity
	p.setGravity(0,0,-10)

	return turtle

turtle = setupWorld()
# enable real time simulation
# if not his, robots will move a lot faster
p.setRealTimeSimulation(1)

forward=0
turn=0

keyBoardEvents = []
posAndOrientation = []
simulate = False

def replay(kb):
	turtle = setupWorld()
	# need to wait for the boxes to be stable and then continue
	print("state restored " + str(len(kb)))
	# p.stepSimulation()
	# p.setTimeStep(1./60.)

	# replay all keyboard events
	for pair in kb:
		print(pair)
		time.sleep(1./100)
		# p.setJointMotorControl2(turtle,0,p.VELOCITY_CONTROL,targetVelocity=pair[0],force=1000)
		# p.setJointMotorControl2(turtle,1,p.VELOCITY_CONTROL,targetVelocity=pair[1],force=1000)
		p.resetBasePositionAndOrientation(turtle, pair[0], pair[1])
	
	# stop the robot
	# p.setJointMotorControl2(turtle,0,p.VELOCITY_CONTROL,targetVelocity=0,force=1000)
	# p.setJointMotorControl2(turtle,1,p.VELOCITY_CONTROL,targetVelocity=0,force=1000)


while (1):

	time.sleep(1./240.)
	keys = p.getKeyboardEvents()

	if p.B3G_RETURN in keys and keys[p.B3G_RETURN] & p.KEY_WAS_TRIGGERED:
		simulate = True
		# replay(keyBoardEvents.copy())
		replay(posAndOrientation.copy())

	if not simulate: 

		leftWheelVelocity=0
		rightWheelVelocity=0
		speed=10
		# turn = 0
		# forward = 0
		
		for k,v in keys.items():

			if (k == p.B3G_RIGHT_ARROW and (v&p.KEY_WAS_TRIGGERED)):
				turn = -0.5
			if (k == p.B3G_RIGHT_ARROW and (v&p.KEY_WAS_RELEASED)):
				turn = 0
			if (k == p.B3G_LEFT_ARROW and (v&p.KEY_WAS_TRIGGERED)):
				turn = 0.5
			if (k == p.B3G_LEFT_ARROW and (v&p.KEY_WAS_RELEASED)):
				turn = 0

			if (k == p.B3G_UP_ARROW and (v&p.KEY_WAS_TRIGGERED)):
				forward=1
			if (k == p.B3G_UP_ARROW and (v&p.KEY_WAS_RELEASED)):
				forward=0
			if (k == p.B3G_DOWN_ARROW and (v&p.KEY_WAS_TRIGGERED)):
				forward=-1
			if (k == p.B3G_DOWN_ARROW and (v&p.KEY_WAS_RELEASED)):
				forward=0

		rightWheelVelocity+= (forward+turn)*speed
		leftWheelVelocity += (forward-turn)*speed

		keyBoardEvents.append([leftWheelVelocity, rightWheelVelocity])
		# rightWheelVelocity==0 and leftWheelVelocity==0 and 
		# if not (rightWheelVelocity==0 and leftWheelVelocity==0):
		# 	keyBoardEvents.append([leftWheelVelocity, rightWheelVelocity])
		
		print("[" + str(leftWheelVelocity) + ", " + str(rightWheelVelocity) + "]")
		p.setJointMotorControl2(turtle,0,p.VELOCITY_CONTROL,targetVelocity=leftWheelVelocity,force=1000)
		p.setJointMotorControl2(turtle,1,p.VELOCITY_CONTROL,targetVelocity=rightWheelVelocity,force=1000)

		pos, ori = p.getBasePositionAndOrientation(turtle)
		temp = []
		temp.append(pos)
		temp.append(ori)
		posAndOrientation.append(temp)
```
